### Q3: Compare and contrast mutable and immutable objects in Python with examples

# Immutable example: int, str, tuple
a = 10
b = a
a = 20
print(b)  # b remains 10

# Mutable example: list, dict
lst1 = [1, 2, 3]
lst2 = lst1
lst1.append(4)

### Q4: Discuss the different types of operators in Python and provide examples of how they are used

# Arithmetic Operators
a, b = 10, 3
print(a + b, a - b, a * b, a / b, a % b, a ** b)

# Comparison Operators
print(a == b, a != b, a > b, a <= b)

# Logical Operators
print(True and False, True or False, not True)

# Assignment Operators
x = 5
x += 2  # x = x + 2
print(x)

# Bitwise Operators
print(a & b, a | b, a ^ b)

# Membership Operators
print(3 in [1, 2, 3])

# Identity Operators
print(a is b, a is not b)

### Q7: Describe the different types of loops in Python and their use cases with examples

# For loop - used for iterating over a sequence
for i in range(5):
    print("For loop:", i)

# While loop - used when the number of iterations is not known
count = 0
while count < 3:
    print("While loop:", count)
    count += 1

# Loop control: break, continue, pass
for i in range(5):
    if i == 3:
        break  # exits loop
    print("Break example:", i)

for i in range(5):
    if i == 2:
        continue  # skips current iteration
    print("Continue example:", i)
